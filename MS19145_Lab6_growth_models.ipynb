```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rList=[1,2,2.2,2.5,3,3.2,3.5,3.8]
fig=plt.figure(figsize=(20,10))
for r in rList:
    tList=np.arange(0,10,0.1)
    Nt=[]
    N0=8
    for t in tList:
        Nt.append(np.log(N0*np.exp(r*t)))
    plt.plot(tList,Nt,label="r="+str(r))
plt.title("Exponential model",fontsize=30)
plt.ylabel(r'$ln(N_{t})$',fontsize=30)
plt.xlabel("t",fontsize=30)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=15)
plt.savefig("exponential_model.png", dpi=200)
plt.show()
```

```python
rList=[1,2,2.2,2.5,3,3.2,3.5,3.8]
kList=[100,500,1000]
for k in kList:
    fig=plt.figure(figsize=(20,10))
    for r in rList:
        N0=8
        tList=np.arange(0,15,0.01)
        Nt=[]
        for t in tList:
            Nt.append(k*N0*np.exp(r*t)/(k+N0*(np.exp(r*t)-1)))
        plt.plot(tList,Nt,label="r="+str(r))
    plt.title("Logistic Model(K="+str(k)+")",fontsize=30)
    plt.xlabel("t",fontsize=30)
    plt.ylabel(r'$N_{t}$',fontsize=30)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(fontsize=15)
    plt.savefig("logistic_model{}.png".format(k), dpi=200)
    plt.show()
```

<div style="text-align:left; margin:50px;">
    Q) Does population size reach an equilibrium?<br><br>
       Yes, the population size reaches an equilibrium.<br><br>
    Q) What is the value of equilibrium point?<br><br>
    At K=100,<br><br>
        $\qquad$$Nt_{eq}$ = 100<br><br>
    At K=500,<br><br>
        $\qquad$$Nt_{eq}$ = 500<br><br>
    At K=1000,<br><br>
        $\qquad$$Nt_{eq}$ = 1000<br><br>
    Q) How do r and K effect equilibrium?<br><br>
    K is the carrying capacity of the environment. At equilibrium, the number of individuals is equal to K. r determines 
    how quickly equilibrium is reached. For a given value of K, the higher value of r lesser time it takes to reach 
    equilibrium. Increasing K increases the time taken to reach equilibrium.
    </div>

```python
rList=[0.5,1,2,2.2,2.5,3]
kList=[100,500,1000]
for k in kList:
    fig=plt.figure(figsize=(25,10))
    for r in rList:
        tList=np.arange(0,100,1)
        Nt=[8]
        for t in tList[1:]:
            Nt.append(Nt[-1]*np.exp(r*(1-(Nt[-1]/k))))
        plt.plot(tList,Nt,label="r="+str(r))
    plt.title("Robert May Model(K="+str(k)+")",fontsize=30)
    plt.xlabel("Number of Generations",fontsize=30)
    plt.ylabel(r'$N_{t}$',fontsize=30)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.legend(fontsize=15)
    plt.savefig("robert_may_model{}.png".format(k), dpi=200)
    plt.show()
```

```python
rList=np.arange(0,3.83,0.01)
kList=[100,500,1000]
for k in kList:
    fig=plt.figure(figsize=(20,10))
    for r in rList:
        tList=np.arange(0,1000,1)
        Nt=[8]
        rs=[r]
        for t in tList[1:]:
            Nt.append(Nt[-1]*np.exp(r*(1-(Nt[-1]/k))))
            rs.append(r)
        plt.scatter(rs[900:1000],Nt[900:1000],color="b")
    plt.title("Robert May Model(k="+str(k)+")",fontsize=30)
    plt.xlabel("r",fontsize=30)
    plt.ylabel(r'$N_{eq}$',fontsize=30)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.savefig("bifurcation_diagram{}.png".format(k), dpi=200)
    plt.show()
```
